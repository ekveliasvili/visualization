Python code for the plot is below:

```python
# download libraries

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('/Users/evgeniakveliashvili/Desktop/DSI_projects/visualization/02_activities/assignments/short-term-rental-registrations-data.csv')
#print (dataset.head())
dataset['property_type'].nunique()
dataset.head()
```

```python
# add random seed to ensure reproducibility 
np.random.seed(123)
property_type_count = dataset['property_type'].value_counts()
x = property_type_count.index
y = property_type_count.values
fig, ax = plt.subplots(figsize = (10,5))
ax.bar(x,y)
ax.set_title('Number of short-term rental registrations by property type', color = 'indigo', size = 16, fontweight = 'bold')
ax.set_xlabel('Type of property', color = 'indigo', size = 12, fontweight = 'bold')
ax.set_ylabel('Number of registrations', color = 'indigo', size = 12, fontweight= 'bold')
ax.set_xticklabels(x, rotation = 45, ha = 'right', size = 12)
```
