```python
# create data 
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

# create a histogram from the dataset above
fig, ax = plt.subplots(figsize = (5,3))
ax.hist (y)
```

```python
# create a plot graph and add title and labels 
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title ("Total growth over time")
ax.set_xlabel("Total growth")
ax.set_ylabel ("Years since start")
```

```python
# font dictionary
font1 = {'family' : 'sans-serif', 'color' : 'purple', 'size' :20}
font2 = {'family' : 'monospace', 'color' : 'orange', 'size': 10}

# create a plot graph and use font dictionaries
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title ("Total growth over time", fontdict=font2)
ax.set_xlabel("Total growth", fontdict=font1)
ax.set_ylabel ("Years since start", fontdict=font2)
```

```python
#create a scatter plot diagram
fig, ax = plt.subplots (figsize = (5,3))
ax.scatter (x,y,
marker = 'd',
color = "indigo")
```

```python
# create a plot graph
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y,
        marker = "*",
        color = "orange",
        linestyle = "--",
        linewidth = 2)
                       
```

```python
# create a grid graph

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y,
        marker = "*",
        color = "orange",
        linestyle = "--",
        linewidth = 2,
        markersize = 12,
        markeredgecolor = "#fa9359",
        markerfacecolor = "black")
ax.grid( axis = "x", color = "blue", linewidth = 2, linestyle = "dotted")
```

```python
#create dataset for the graph 
height = [3, 12, 5, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')

df = pd.DataFrame({'height': height,
                   'bars': bars})

# create a list of colors for the plot
color = ("blue", "red", "orange", "yellow", "green"
         ) 

# make a graph and add title and labels
fig, ax = plt.subplots()
ax.bar(df['bars'], df["height"],
       color = color)
ax.set_title("My bar diagram")
ax.set_xlabel("years")
ax.set_ylabel(" Average age")
```

```python
# create data set with original data 
name = ["red", "green", "blue","orange"]
size_of_groups=[12,11,3,30]

# Create a pieplot
plt.pie(size_of_groups, labels = name)

# add a circle at the center to transform it in a donut chart
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```
