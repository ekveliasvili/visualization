```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import PIL
import requests
```

```python
# create dataset for building the plot
np.random.seed (613)
x = np.arange (50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(1,100,50)
```

```python
# create a simple plot using the dataset above
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc = "upper left")                       
```

```python
# create a simple plot and add label features like font, shadow and columns
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc = "upper left", # customize labels
          frameon = True,
          fontsize = 9, # add fontsize
          ncol = 2, 
          shadow = False)    
```

```python
fig, ax = plt.subplots(figsize =(5,3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc = "upper left",
             bbox_to_anchor = (1,1)) # move the legend outside of the plot
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = "lower right")
# add text on the plot nand customize it
ax.text(10,95, "This is very important",
        ha = 'left',
        color = 'purple',
        size = 12)

```

```python
# create new dataset for plot

np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(["Luffy", "Zoro", "Usopp", "Sanji", "Nami"])
y2 = np.array([110,180,240,99,220])
```

```python
# create two plots next to each other
fig, (ax1, ax2) = plt.subplots(ncols = 2,
                               nrows = 1,
                               figsize = (7,3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
```

```python
# create two plots next to each other
fig, (ax1, ax2) = plt.subplots(ncols = 2,
                               nrows = 1,
                               figsize = (7,3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
ax1.set_xlabel ("label 1", size = 10, color = "green")
ax1.set_ylabel ("label 2", size = 10, color = "green")
ax1.set_title ("Fig 1")

ax2.set_xlabel("better label", size = 8, color = "red", ha = 'center')
ax2.set_ylabel("number of reads", size = 8, color = "red")
ax2.set_title ("Fig 2")
```

```python
# create several plots in one figure
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize = (7,4),
                                    layout = "constrained")
someaxes['ax1'].scatter (x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)
someaxes['ax1'].set_xlabel("A big label", fontsize = 18)
someaxes["ax2"].set_xlabel ("Anohter label", fontsize = 18)
someaxes["ax3"].set_xlabel ("a realy big big one", fontsize = 18)
```

```python
# create two graphs on one axes
x = np.array (["Luffy", "Zoro", "Nami", "One", "Two"])
y1 = np.array([110,180,240,99,220])
y2 = np.array([170,100,99,38,87])
fig, ax = plt.subplots(figsize = (5,3))
ax.bar(x,y1, color = "indigo")
ax.plot(x,y2, color = "red")
```

```python
# add standar erorr on the plot
y2_sd = np.std(y2)
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y2, color = "red")
ax.errorbar(x,y2,
            yerr= y2_sd,
            fmt = "none",
            ecolor = "green",
            elinewidth = 3,
            capsize = 4,
            capthick=4,
            errorevery=3)
```
