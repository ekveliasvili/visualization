```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import scipy
import PIL
import requests
```

```python
tips = sns.load_dataset ("tips")
print(tips)
```

```python
# build simple linecode
sns.set_style('whitegrid')
tipgraph = sns.lineplot(data = tips,
             x = 'total_bill',
             y = 'tip')
tipgraph.set(title = 'Tips vs Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip amoiunt($)')
```

```python
fig = plt.subplots(figsize=(10,3))
```

```python
tipgraph = sns.lineplot(data = tips,
                        x = "total_bill",
                        y = "tip",
                        color = 'hotpink',
                        linestyle = "--",
                        linewidth = 3,
                        marker = "o",
                        markerfacecolor = "pink")

tipgraph.set(title = 'Tips vs Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip amount($)')
```

```python
tips = sns.load_dataset ("tips")
tipgraph = sns.scatterplot(data = tips,
                           x = "total_bill",
                           y= "tip",
                           style = "time",
                           hue = 'day',
                           palette = ['purple', 'hotpink', 'deepskyblue', 'yellowgreen'])
```

```python
sns.pairplot(data = tips, hue = 'day')
```

```python

import plotly.graph_objects as go
x1 = np.array(['Luffy', 'Zoro', 'Nami', 'Usopp', 'Sanjil'])
y1 = np.array([110, 180, 240, 99, 220])
```

```python
graph = go.Figure()
graph.add_trace(go.Scatter(x=x1, y=y1,
                           mode = 'markers',
                           marker = dict(
                               size = 15,
                               color = 'hotpink',
                               opacity = 1)))
graph.update_layout(title = 'Pirate stores',
                    xaxis_title = 'Pirates',
                    yaxis_title ='Score',
                    width = 700,
                    height = 700)
```

```python
from wordcloud import WordCloud
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv')
```

```python
text = " ".join(each for each in df.quote)
wordcloud = WordCloud(background_color = 'white',
                      colormap = 'inferno').generate(text)
fig, ax = plt.subplots(figsize= (7,3))
ax.imshow(wordcloud)
ax.axis('off')
```

```python
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
A = set(["apple", 'banana', 'watermelon'])
B = set (['pumpkin', 'blueberry', 'apple', 'key limme'])
```

```python
diagram = venn2([A,B],
                set_labels = ('Fruits', 'Pires'),
                set_colors = ('blue', 'red'),
                alpha = 0.5)
diagram.get_label_by_id('10').set_text("\n".join(A-B))
diagram.get_label_by_id('11').set_text("\n".join(A&B))
diagram.get_label_by_id('01').set_text("\n".join(B-A))
```
